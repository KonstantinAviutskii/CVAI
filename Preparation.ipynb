```python
import os
import shutil
from tensorflow.keras.preprocessing.image import load_img, img_to_array, save_img
from tensorflow.keras.applications.vgg16 import preprocess_input

# Paths to the folders in Google Drive
dataset_dir = '/content/drive/MyDrive/CVAI/Task1/chest_xray/test'
processed_dir = '/content/drive/MyDrive/CVAI/Task1/chest_xray/299_test'  # Destination for processed images

# Remove existing processed folder if exists
if os.path.exists(processed_dir):
    shutil.rmtree(processed_dir)
os.makedirs(processed_dir, exist_ok=True)

# Loop through class folders
for class_name in os.listdir(dataset_dir):
    class_path = os.path.join(dataset_dir, class_name)
    if os.path.isdir(class_path):
        # Create folder in processed dataset
        processed_class_path = os.path.join(processed_dir, class_name)
        os.makedirs(processed_class_path, exist_ok=True)

        # Process each image
        for filename in os.listdir(class_path):
            if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
                img_path = os.path.join(class_path, filename)
                try:
                    # Load, resize, convert to array
                    img = load_img(img_path, target_size=(299, 299))
                    img_array = img_to_array(img)
                    # Preprocess for VGG16
                    img_array = preprocess_input(img_array)
                    # Save the preprocessed image
                    save_path = os.path.join(processed_class_path, filename)
                    save_img(save_path, img_array)
                except Exception as e:
                    print(f"Error processing {img_path}: {e}")

print("Dataset preprocessing complete.")
```
